```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("dataset")

# Calculate the absolute relative deviation and the percentile rank of samples
df["Desvio Relativo"] = abs((df["Spl2"] - df["Spl1"]) / df["Spl1"])
df["Percentile Rank"] = df["Desvio Relativo"].rank(pct=True)

fig, ax = plt.subplots(figsize=(10, 6))

#Ploting the relative deviation graph as a percentage from 0 to 100
ax.scatter(df["Percentile Rank"]*100, df["Desvio Relativo"]*100)
ax.set_xlabel("Percentile Rank (%)")
ax.set_ylabel("Absolute relative deviation (%)")
ax.set_ylim(0, 100)

#Create a redline for the percentile rank and the Absolute relative deviation
ax.axvline(x=90, color="red")
ax.axhline(y=10, color="red")
ax.set_xticks(np.arange(0, 101, 10))
ax.set_yticks(np.arange(0, 101, 10))

plt.show()
```
